#### Changing date fromat from %d/%m/%Y to %Y-%m-%d as all other datasets follow this format

```python
from datetime import datetime
gp['Date'] = gp['Date'].dt.strftime('%Y-%m-%d')
gp.head()
```

```python
gp['Date'] = pd.to_datetime(gp['Date'])
gp.head()
```

###Loading Repo Rate Dataset

```python
rr = rr[['Date', 'RepoRate']]
rr.head()
```

```python
rr.info()
```

```python
rr['Date'] = pd.to_datetime(rr['Date'])
rr.info()
```

Droping 1st row as it contains wrong value from previous date

```python
rr.drop(index=0, inplace=True)
rr.head()
```

###Loading USD INR dataset

```python
ui = pd.read_csv('/content/usdInrDataset.csv')
ui.head()
```

## Merge all datasets into one dataset

```python
df = gp.merge(br, on='Date', how='left')
df.head()
```

```python
df.head(35)
```

```python
df = df.merge(co, on='Date', how='left')
df.head(35)
```

```python
df = df.merge(rr, on='Date', how='left')
df.head(35)
```

```python
df = df.merge(ui, on='Date', how='left')
df.head(35)
```

Adding values from April 1st 2001 to BankRate and RepoRate as the goldPrice dataset is starting from April 2nd 2001

```python
df['BankRate'][0] = 7
df['RepoRate'][0] = 9
df.head()
```

### Fill NA with previous values as the BankRate, Reporate are constant untill new rates are released

```python
df.fillna(method='ffill', inplace=True)
df.head(35)
```

```python
df.tail()
```

```python
df['GoldPrice'] = df['GoldPrice'].apply(lambda x: x.replace(',',''))
df.head()
```

```python
df['GoldPrice'] = pd.to_numeric(df['GoldPrice'])
```

```python
df.tail()
```

```python
def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
```

```python
x = set()
df['UsdInr'].apply(lambda l : x.add(l) if not is_float(l) else 'num')
x
```

```python
df.describe()
```

CrudeOil rate and USDINR have '.' instead of NAN, so replace the '.' with NAN

```python
df['CrudeOilRate'] = df['CrudeOilRate'].apply(lambda l: l if is_float(l) else None)
```

```python
df.isnull().sum()
```

```python
df['UsdInr'] = df['UsdInr'].apply(lambda l: l if is_float(l) else None)
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.info()
```

```python
df['CrudeOilRate'] = pd.to_numeric(df['CrudeOilRate'])
```

```python
df['UsdInr'] = pd.to_numeric(df['UsdInr'])
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
x = set()
df['CrudeOilRate'].apply(lambda l : x.add(l) if l <0 else '')
x
```
